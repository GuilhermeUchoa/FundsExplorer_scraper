```python
Newdf = df[(df['Setor']=='Logística') & 
           (df['Dividend Yield']>=0.5) &
           (df['Quantidade Ativos'] >= 5) &
           (df['Preço Atual'] <= df['VPA']*1.1) &
           (df['Preço Atual'] >= df['VPA']*0.80)]
           #(df['Rentab. Período'] >= 0)
          
        
Newdf['Mil_investido'] = (1000//Newdf['Preço Atual'])*Newdf['Dividendo']
```

```python
Newdf.sort_values(by='Mil_investido',ascending=False)[['Código do fundo','Setor','Preço Atual','VPA','Dividendo',
                                                       'Dividend Yield','Vacância Física','Quantidade Ativos','Rentab. Período',
                                                       'Mil_investido']]
```
